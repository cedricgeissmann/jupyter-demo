## Aufgabe Schritt für Schritt vorrechnen

Bestimmen Sie den Punkt der auf der Gerade $f(x) = y = 2 \cdot x - 1$, der dem Ursprung am nächsten liegt.

```python
# Skizze der Geraden

x = symbols('x')
f = 2 * x - 1
plot(f, xlim=[-5, 5], ylim=[-5,5])
```

### Der Abstand $d$

Den Abstand zum Ursprung können wir als $d$ definieren:

$$ \begin{align*}
    d^2 &= x^2 + f(x)^2 \\
    d^2 &= x^2 + (2 \cdot x - 1)^2
    \end{align*}
$$

Weil $d$ und $d^2$ das Maximum an der gleichen Stelle haben, und wir nur an der Stelle $x$ interessiert sind, können wir vereinfacht mit $d$ rechnen.

$$
    d = x^2 + (2 \cdot x - 1)^2
$$

```python
# d definieren und Skizzieren
d = x**2 + (2*x - 1)**2

plot(f, d, xlim=[-5, 5], ylim=[-5,5])
```

### $d$ vereinfachen und den Scheitelpunkt berechnen

$$
    S = \left(- \frac{b}{2\cdot a}, \frac{-b^2 + 4 \cdot a \cdot c}{4 \cdot a} \right)
$$

```python
print(d)
# d vereinfachen
d = expand(d)
print(d)
```

```python
# x-Koordinate von S berechnen

x_s = - (-4)/(2*5)
print(x_s)
```

```python
# x_s in f einsetzen
y_s = 2 * x_s -1

print(x_s, y_s)
```

### Lösung

Der Punkt auf der Geraden $f$, der dem Ursprung am nächsten ist, ist der Punkt $P = (0.4, -0.2)$.

### Beispiel Scheitelpunkt

Wir können eine Funktion zur Verfügung stellen, die den Scheitelpunkt berechnet.

```python
def scheitelpunkt(a=1, b=1, c=0):
    x_s = - b/(2*a)
    y_s = (-(b**2) + 4*a*c)/(4*a)
    print(f'Der Scheitelpunkt ist ({x_s}, {y_s})')
    
scheitelpunkt(a=5, b=-4, c=1)
```

# Algebra mit Python

Python hat eine Bibliothek `sympy` mit der symbolisches rechnen in Python möglich ist.

- Ziemlich umfangreich
- Kann erweitert werden
- Workflows können vereinfacht werden

```python
from sympy import *

a, x = symbols('a x')

f = a*x**2 - 2*x + 4

integrate(f, x)
```

# Symbolisches integrieren

Mit sympy können wir auch integrieren.

```python
a, x = symbols('a x')
f = a * (1 - 0.25*x**2)

nullstellen = solve(f, x)
print(f'Die Nullstellen von {f} sind: {nullstellen}')
```

```python
F = integrate(f, (x, -2, 2))
area = 8
solve(F-area, a)
```
